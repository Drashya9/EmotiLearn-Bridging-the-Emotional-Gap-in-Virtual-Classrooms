```python
import os
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA, TruncatedSVD
from sklearn.preprocessing import MinMaxScaler

# === CONFIG ===
input_dir = r"D:/DAiSEE/DAiSEE/transformer_1000/OpenFace_Features"        
output_dir = r"D:/DAiSEE/DAiSEE/transformer_1000/Preprocessed_Features"    
os.makedirs(output_dir, exist_ok=True)

use_pca = True 

for video_id in os.listdir(input_dir):
    video_folder = os.path.join(input_dir, video_id)
    if not os.path.isdir(video_folder):
        continue

    # Look for processed file
    csv_candidates = [f for f in os.listdir(video_folder) if f.endswith(".csv")]
    if not csv_candidates:
        print(f"No CSV found for {video_id}")
        continue

    csv_path = os.path.join(video_folder, csv_candidates[0])
    df = pd.read_csv(csv_path)
    df.columns = df.columns.str.strip()  

    # Ensure required columns exist 
    if not set(['frame', 'confidence', 'success']).issubset(df.columns):
        print(f"Skipping {video_id}: required columns missing")
        continue

    #  Data Selection
    df = df[df['success'] == 1]
    if df.empty:
        print(f"Skipping {video_id}: no successful frames")
        continue

    df = df.loc[df.groupby("frame")["confidence"].idxmax()]

    # Select only feature columns 
    meta_cols = ['frame', 'face_id', 'timestamp', 'confidence', 'success']
    feature_cols = [col for col in df.columns if col not in meta_cols]
    X = df[feature_cols].values

    if X.shape[0] < 300:
        pad_len = 300 - X.shape[0]
        X = np.pad(X, ((0, pad_len), (0, 0)), mode='edge')
    else:
        X = X[:300, :]


    #PCA to reduce to (300, 300)
    reducer = PCA(n_components=300) if use_pca else TruncatedSVD(n_components=300)
    X_reduced = reducer.fit_transform(X)

    # Min-Max Normalize to [0, 1]
    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X_reduced)

    # Save the matrix as .npy
    output_path = os.path.join(output_dir, f"{video_id}.npy")
    np.save(output_path, X_scaled)
    print(f" Saved: {output_path} | shape: {X_scaled.shape}")

print("🎉 All videos processed and saved.")
```

```python
import os
import numpy as np
import pandas as pd
from imblearn.over_sampling import SMOTE
from collections import Counter

features_dir = r"D:/DAiSEE/DAiSEE/transformer_1000/Preprocessed_Features"
label_csv_path = r"D:/DAiSEE/DAiSEE/transformer_1000/filtered_video_labels.csv"
output_dir = r"D:/DAiSEE/DAiSEE/transformer_1000/smote_balanced"
os.makedirs(output_dir, exist_ok=True)

emotion_columns = ['Boredom', 'Engagement', 'Confusion', 'Frustration']

# Load filtered label CSV
df = pd.read_csv(label_csv_path)
df.columns = df.columns.str.strip()
df['video_id'] = df['video_id'].astype(str).str.strip()

# Load feature vectors and average them
X_all = []
valid_ids = []

for _, row in df.iterrows():
    vid = row['video_id']
    npy_path = os.path.join(features_dir, f"{vid}.npy")

    if os.path.exists(npy_path):
        matrix = np.load(npy_path)

        # Pad or trim to 300 frames
        if matrix.shape[0] < 300:
            pad_len = 300 - matrix.shape[0]
            matrix = np.pad(matrix, ((0, pad_len), (0, 0)), mode='edge')
        else:
            matrix = matrix[:300, :]

        if matrix.shape[1] != 300:
            print(f"Skipping {vid}: unexpected shape {matrix.shape}")
            continue

        avg_feature = matrix.mean(axis=0)  # (300,)
        X_all.append(avg_feature)
        valid_ids.append(vid)

# Final DataFrame filtering and feature matrix
df = df[df['video_id'].isin(valid_ids)].reset_index(drop=True)
X_all = np.array(X_all)

# SMOTE per emotion
for emotion in emotion_columns:
    y = df[emotion].astype(int).values
    print(f"\n[Before SMOTE] {emotion} distribution:")
    print(pd.Series(y).value_counts().sort_index())

    class_counts = Counter(y)
    min_class_count = min(class_counts.values())
    k_neighbors = min(min_class_count - 1, 5)

    if k_neighbors < 1:
        print(f"Not enough samples for SMOTE on {emotion}. Skipping...")
        continue

    smote = SMOTE(random_state=42, k_neighbors=k_neighbors)
    X_resampled, y_resampled = smote.fit_resample(X_all, y)

    print(f"[After SMOTE] {emotion} distribution:")
    print(pd.Series(y_resampled).value_counts().sort_index())

    # Save balanced data
    np.save(os.path.join(output_dir, f"{emotion}_features.npy"), X_resampled)
    np.save(os.path.join(output_dir, f"{emotion}_labels.npy"), y_resampled)

print("\nAll emotion classes balanced using SMOTE and saved.")
```

```python
import os
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset, random_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Transformer Model 
class VideoTransformer(nn.Module):
    def __init__(self, input_dim=300, seq_len=1, num_classes=4, d_model=256, nhead=8, num_layers=4):
        super(VideoTransformer, self).__init__()
        self.input_fc = nn.Linear(input_dim, d_model)
        self.pos_embedding = nn.Parameter(torch.randn(1, seq_len, d_model))
        encoder_layer = nn.TransformerEncoderLayer(d_model=d_model, nhead=nhead, dim_feedforward=512, dropout=0.1, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.cls_head = nn.Sequential(
            nn.Linear(d_model, 128),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):  # x: (B, T, F)
        x = self.input_fc(x)
        x = x + self.pos_embedding
        encoded = self.transformer_encoder(x)
        pooled = encoded.mean(dim=1)  # mean pooling
        return self.cls_head(pooled)

# Training Function 
def train_on_smote_data(smote_dir, emotion):
    print(f"\nTraining Transformer on {emotion.upper()}")

    # Load Data 
    X = np.load(os.path.join(smote_dir, f"{emotion}_features.npy"))  # shape: (N, 300)
    y = np.load(os.path.join(smote_dir, f"{emotion}_labels.npy"))

    # Convert to sequence shape for Transformer: (N, 1, 300)
    X_tensor = torch.tensor(X[:, None, :], dtype=torch.float32)
    y_tensor = torch.tensor(y, dtype=torch.long)

    dataset = TensorDataset(X_tensor, y_tensor)
    train_size = int(0.8 * len(dataset))
    test_size = len(dataset) - train_size
    train_set, test_set = random_split(dataset, [train_size, test_size])
    train_loader = DataLoader(train_set, batch_size=8, shuffle=True)
    test_loader = DataLoader(test_set, batch_size=8)

    # Model Setup
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = VideoTransformer(seq_len=1).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

    # Training Loop 
    best_acc = 0
    for epoch in range(30):  # adjust epochs as needed
        model.train()
        total_loss, preds, labels = 0, [], []

        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad()
            out = model(xb)
            loss = criterion(out, yb)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()
            preds.extend(out.argmax(dim=1).cpu().numpy())
            labels.extend(yb.cpu().numpy())

        acc = accuracy_score(labels, preds) * 100
        print(f"Epoch {epoch+1}: Loss={total_loss/len(train_loader):.4f}, Accuracy={acc:.2f}%")

        if acc > best_acc:
            best_acc = acc
            torch.save(model.state_dict(), f"best_model_{emotion}.pth")

    # Final Evaluation
    model.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for xb, yb in test_loader:
            xb, yb = xb.to(device), yb.to(device)
            out = model(xb)
            all_preds.extend(out.argmax(dim=1).cpu().numpy())
            all_labels.extend(yb.cpu().numpy())

    test_acc = accuracy_score(all_labels, all_preds) * 100
    print(f"\nFinal Test Accuracy for {emotion}: {test_acc:.2f}%")

    # Confusion Matrix
    cm = confusion_matrix(all_labels, all_preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Very Low", "Low", "High", "Very High"])
    disp.plot(cmap='Blues')
    plt.title(f"{emotion.capitalize()} - Confusion Matrix")
    plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset, random_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn.utils.class_weight import compute_class_weight


# === Model Definition ===
class VideoTransformer(nn.Module):
    def __init__(self, input_dim=300, seq_len=300, num_classes=4, d_model=256, nhead=8, num_layers=4):
        super(VideoTransformer, self).__init__()
        self.input_fc = nn.Linear(input_dim, d_model)
        self.pos_embedding = nn.Parameter(torch.randn(1, seq_len, d_model))
        encoder_layer = nn.TransformerEncoderLayer(d_model=d_model, nhead=nhead, dim_feedforward=512, dropout=0.1, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.cls_head = nn.Sequential(
            nn.Linear(d_model, 128),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):  # (B, T, F)
        x = self.input_fc(x)
        x = x + self.pos_embedding
        encoded = self.transformer_encoder(x)
        pooled = encoded.mean(dim=1)
        return self.cls_head(pooled)

# === Training Without SMOTE ===
def train_from_raw_data(features_dir, label_csv_path, emotion, output_dir="outputs"):
    print(f"\nTraining on original data for: {emotion.upper()}")
    os.makedirs(output_dir, exist_ok=True)

    # === Load and filter label CSV
    df = pd.read_csv(label_csv_path)
    df.columns = df.columns.str.strip()
    df['video_id'] = df['ClipID'].astype(str).str.replace(".avi", "").str.replace(".mp4", "").str.strip()

    X_all, y_all = [], []
    for _, row in df.iterrows():
        vid = row['video_id']
        label = row[emotion]
        npy_path = os.path.join(features_dir, f"{vid}.npy")
        if os.path.exists(npy_path):
            arr = np.load(npy_path)
            if arr.shape[0] < 300:
                pad_len = 300 - arr.shape[0]
                arr = np.pad(arr, ((0, pad_len), (0, 0)), mode='edge')
            else:
                arr = arr[:300, :]

            if arr.shape[1] != 300:
                continue

            X_all.append(arr)
            y_all.append(label)

    print(f"Total samples used: {len(X_all)}")

    X_tensor = torch.tensor(np.array(X_all), dtype=torch.float32)
    y_tensor = torch.tensor(np.array(y_all), dtype=torch.long)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
    # Compute class weights for removing smae type predictions and punishing
    class_weights = compute_class_weight(class_weight='balanced', classes=np.unique(y_all), y=y_all)
    class_weights_tensor = torch.tensor(class_weights, dtype=torch.float32).to(device)


    dataset = TensorDataset(X_tensor, y_tensor)
    train_size = int(0.8 * len(dataset))
    test_size = len(dataset) - train_size
    train_set, test_set = random_split(dataset, [train_size, test_size])
    train_loader = DataLoader(train_set, batch_size=8, shuffle=True)
    test_loader = DataLoader(test_set, batch_size=8)

    # === Model
    model = VideoTransformer(seq_len=300).to(device)
    criterion = nn.CrossEntropyLoss(weight=class_weights_tensor)
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

    best_acc = 0
    for epoch in range(10):
        model.train()
        total_loss, preds, labels = 0, [], []
        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad()
            out = model(xb)
            loss = criterion(out, yb)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
            preds.extend(out.argmax(dim=1).cpu().numpy())
            labels.extend(yb.cpu().numpy())

        acc = accuracy_score(labels, preds) * 100
        print(f"Epoch {epoch+1}: Loss={total_loss/len(train_loader):.4f}, Accuracy={acc:.2f}%")
        if acc > best_acc:
            best_acc = acc
            torch.save(model.state_dict(), os.path.join(output_dir, f"best_model_raw_{emotion}.pth"))

    # === Evaluation
    model.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for xb, yb in test_loader:
            xb, yb = xb.to(device), yb.to(device)
            out = model(xb)
            all_preds.extend(out.argmax(dim=1).cpu().numpy())
            all_labels.extend(yb.cpu().numpy())

    test_acc = accuracy_score(all_labels, all_preds) * 100
    print(f"\nFinal Test Accuracy for {emotion}: {test_acc:.2f}%")

    cm = confusion_matrix(all_labels, all_preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Very Low", "Low", "High", "Very High"])
    disp.plot(cmap='Blues')
    plt.title(f"{emotion.capitalize()} - Confusion Matrix")
    fig_path = os.path.join(output_dir, f"confusion_matrix_raw_{emotion}.png")
    plt.savefig(fig_path)
    print(f"Saved confusion matrix to {fig_path}")
    plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset, random_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.utils.class_weight import compute_class_weight
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt


# Configuration
FEATURES_DIR = "D:/DAiSEE/DAiSEE/transformer_1000/Preprocessed_Features"
LABEL_CSV = "D:/DAiSEE/DAiSEE/transformer_1000/filtered_video_labels.csv"
BINARY_OUTPUT_DIR = "D:/DAiSEE/DAiSEE/transformer_1000/binary_preprocessed"

os.makedirs(BINARY_OUTPUT_DIR, exist_ok=True)

class VideoTransformer(nn.Module):
    def __init__(self, input_dim=300, seq_len=300, num_classes=2, d_model=256, nhead=8, num_layers=4):
        super(VideoTransformer, self).__init__()
        self.input_fc = nn.Linear(input_dim, d_model)
        self.pos_embedding = nn.Parameter(torch.randn(1, seq_len, d_model))
        encoder_layer = nn.TransformerEncoderLayer(d_model=d_model, nhead=nhead, dim_feedforward=512, dropout=0.1, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.cls_head = nn.Sequential(
            nn.Linear(d_model, 128),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        x = self.input_fc(x)
        x = x + self.pos_embedding
        encoded = self.transformer_encoder(x)
        pooled = encoded.mean(dim=1)
        return self.cls_head(pooled)

def preprocess_and_save_smote(emotion):
    df = pd.read_csv(LABEL_CSV)
    df.columns = df.columns.str.strip()
    df['video_id'] = df['ClipID'].astype(str).str.replace(".avi", "").str.replace(".mp4", "").str.strip()

    df = df[df[emotion].isin([2, 3])].copy()
    df[emotion] = df[emotion].replace({2: 0, 3: 1}).astype(int)

    X_all, y_all = [], []
    for _, row in df.iterrows():
        vid = row['video_id']
        label = row[emotion]
        npy_path = os.path.join(FEATURES_DIR, f"{vid}.npy")

        if os.path.exists(npy_path):
            arr = np.load(npy_path)
            if arr.shape[0] < 300:
                arr = np.pad(arr, ((0, 300 - arr.shape[0]), (0, 0)), mode='edge')
            else:
                arr = arr[:300, :]

            if arr.shape[1] == 300:
                X_all.append(arr.flatten())
                y_all.append(label)

    X_all = np.array(X_all)
    y_all = np.array(y_all)

    smote = SMOTE(random_state=42)
    X_bal, y_bal = smote.fit_resample(X_all, y_all)
    X_bal = X_bal.reshape(-1, 300, 300)

    np.save(os.path.join(BINARY_OUTPUT_DIR, f"{emotion}_X_smote.npy"), X_bal)
    np.save(os.path.join(BINARY_OUTPUT_DIR, f"{emotion}_y_smote.npy"), y_bal)
    print(f"{emotion}: Saved {len(X_bal)} SMOTE-balanced samples.")

def train_binary_transformer(emotion):
    print(f"\nTraining on {emotion.upper()} (Binary w/ SMOTE)")

    X = np.load(os.path.join(BINARY_OUTPUT_DIR, f"{emotion}_X_smote.npy"))
    y = np.load(os.path.join(BINARY_OUTPUT_DIR, f"{emotion}_y_smote.npy"))

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    X_tensor = torch.tensor(X, dtype=torch.float32)
    y_tensor = torch.tensor(y, dtype=torch.long)

    class_weights = compute_class_weight(class_weight='balanced', classes=np.unique(y), y=y)
    class_weights_tensor = torch.tensor(class_weights, dtype=torch.float32).to(device)

    dataset = TensorDataset(X_tensor, y_tensor)
    train_size = int(0.8 * len(dataset))
    train_set, test_set = random_split(dataset, [train_size, len(dataset) - train_size])
    train_loader = DataLoader(train_set, batch_size=8, shuffle=True)
    test_loader = DataLoader(test_set, batch_size=8)

    model = VideoTransformer(seq_len=300, num_classes=2).to(device)

    
    criterion = nn.CrossEntropyLoss(weight=class_weights_tensor)

    optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
    best_acc = 0

    for epoch in range(10):
        model.train()
        total_loss, preds, labels = 0, [], []
        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad()
            out = model(xb)
            loss = criterion(out, yb)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
            preds.extend(out.argmax(dim=1).cpu().numpy())
            labels.extend(yb.cpu().numpy())

        acc = accuracy_score(labels, preds) * 100
        print(f"Epoch {epoch+1}: Loss={total_loss/len(train_loader):.4f}, Accuracy={acc:.2f}%")
        if acc > best_acc:
            best_acc = acc
            torch.save(model.state_dict(), os.path.join(BINARY_OUTPUT_DIR, f"best_model_binary_{emotion}.pth"))

    model.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for xb, yb in test_loader:
            xb, yb = xb.to(device), yb.to(device)
            out = model(xb)
            all_preds.extend(out.argmax(dim=1).cpu().numpy())
            all_labels.extend(yb.cpu().numpy())

    test_acc = accuracy_score(all_labels, all_preds) * 100
    print(f"Final Test Accuracy for {emotion}: {test_acc:.2f}%")

    cm = confusion_matrix(all_labels, all_preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["High", "Very High"])
    disp.plot(cmap='Blues')
    plt.title(f"{emotion.capitalize()} - Binary Confusion Matrix")
    fig_path = os.path.join(BINARY_OUTPUT_DIR, f"confusion_matrix_binary_{emotion}.png")
    plt.savefig(fig_path)
    print(f"Saved confusion matrix to {fig_path}")
    plt.show()

# Run for all emotions
emotions = ['Frustration','Engagement']
for emo in emotions:
    preprocess_and_save_smote(emo)
    train_binary_transformer(emo)
```
